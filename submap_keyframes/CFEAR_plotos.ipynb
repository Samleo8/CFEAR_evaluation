```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv



sns.set_theme(style="ticks", color_codes=True)
sns.set(style="ticks")

#Categorical
#tips = sns.load_dataset("tips")



df = pd.read_csv('./CFEAR_eval2021-07-15_13:48.csv', sep=r',', header=0)

print(df.columns)


sns_plot = sns.catplot(x=" submap keyframes", y="Transl. Error", hue=" cost type", kind="box", dodge=True, data=df)
plt.ylim(1.0,4)
plt.grid()
plt.show()
sns_plot.savefig('Transl_error_submap_keyframes.eps', format='eps')
sns_plot.savefig('Transl_error_submap_keyframes.png', format='png', dpi=1000)

sns_plot = sns.catplot(x=" submap keyframes", y=" Rot.err. [deg/100m]", hue=" cost type", kind="box", dodge=True, data=df)
plt.ylim(0.3,0.8)
plt.grid()
plt.show()
sns_plot.savefig('rot_error_submap_keyframes.eps', format='eps')
sns_plot.savefig('rot_error_submap_keyframes.png', format='png', dpi=1000)

sns_plot = sns.catplot(x=" submap keyframes", y="RPE [m]", hue=" cost type", kind="box", dodge=True, data=df)
plt.grid()
plt.show()
sns_plot.savefig('rpe_submap_keyframes.eps', format='eps')
sns_plot.savefig('rpe_submap_keyframes.png', format='png', dpi=1000)

sns_plot = sns.catplot(x=" submap keyframes", y=" bias-x [m]", hue=" cost type", kind="box", dodge=True, data=df)
plt.ylim(-0.03,0.03)
plt.grid()
plt.show()
sns_plot.savefig('bias_x_submap_keyframes.eps', format='eps')
sns_plot.savefig('bias_x_submap_keyframes.png', format='png', dpi=1000)

sns_plot = sns.catplot(x=" submap keyframes", y=" ATE [m]", hue=" cost type", kind="box", dodge=True, data=df)
plt.ylim(0,100)
plt.grid()
plt.show()
sns_plot.savefig('ATE_submap_keyframes.eps', format='eps')
sns_plot.savefig('ATE_submap_keyframes.png', format='png', dpi=1000)



```
