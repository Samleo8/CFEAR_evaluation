```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv



sns.set_theme(style="ticks", color_codes=True)
sns.set(style="ticks")

#Categorical
#tips = sns.load_dataset("tips")

colors = ["#E59866", "#D35400", "#873600", "#85C1E9", "#3498DB", "#2874A6"]
colors_two = [ "#D35400", "#3498DB"]
pal=sns.color_palette(colors_two)

df = pd.read_csv('./CFEAR_submap_keyframe_eval_2021-07-15_1348.csv', sep=r',', header=0)

df = df.rename(columns={'Transl. Error': 'Transl. Error[%]'})
df = df.rename(columns={'resolution r': 'resolution r [m]'})

print(df.columns)


sns_plot = sns.catplot(x=" submap keyframes", y="Transl. Error[%]", hue=" cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
plt.title('Transl. Error vs number of submap keyframes')
plt.ylim(1.3,3.8)
plt.grid()
plt.show()
sns_plot.savefig('./output/Transl_error_submap_keyframes.eps', format='eps')
sns_plot.savefig('./output/Transl_error_submap_keyframes.png', format='png', dpi=1500)

sns_plot = sns.catplot(x=" submap keyframes", y=" Rot.err. [deg/100m]", hue=" cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
plt.title('Rotation. Error vs number of submap keyframes')
plt.ylim(0.3,0.82)
plt.grid()
plt.show()
sns_plot.savefig('./output/rot_error_submap_keyframes.eps', format='eps')
sns_plot.savefig('./output/rot_error_submap_keyframes.png', format='png', dpi=1500)

sns_plot = sns.catplot(x=" submap keyframes", y="RPE [m]", hue=" cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
plt.title('Relative Pose Error (RPE)[m] vs submap keyframes')
plt.grid()
plt.show()
sns_plot.savefig('./output/rpe_submap_keyframes.eps', format='eps')
sns_plot.savefig('./output/rpe_submap_keyframes.png', format='png', dpi=1500)

sns_plot = sns.catplot(x=" submap keyframes", y=" bias-x [m]", hue=" cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
plt.title('longitudinal bias vs number of submap keyframes')

plt.grid()
plt.show()
#sns_plot.legend(loc='upper right', ncol=1)
plt.ylim(-0.028,0.02)
#sns_plot.map(plt.axhline, y=0, ls='--', c='red')

#sns_plot.axhline(0)

sns_plot.savefig('./output/bias_x_submap_keyframes.eps', format='eps')
sns_plot.savefig('./output/bias_x_submap_keyframes.png', format='png', dpi=1500)

sns_plot = sns.catplot(x=" submap keyframes", y=" ATE [m]", hue=" cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
plt.title('Absolute Trajectory Error (ATE) vs submap keyframes')
#sns_plot.legend(loc='upper right', ncol=1)
plt.ylim(0,80)
plt.grid()
plt.show()
sns_plot.savefig('./output/ATE_submap_keyframes.eps', format='eps')
sns_plot.savefig('./output/ATE_submap_keyframes.png', format='png', dpi=1500)



```
